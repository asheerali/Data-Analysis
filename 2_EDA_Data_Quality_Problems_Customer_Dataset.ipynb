#### Validating Email

```python
regex = re.compile ("[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
cnt = 0
for value in df.Email:
    if re.search(regex, value):
        print(value, cnt)
    else:
        print('not value')
    cnt+=1
```

```python
regex = re.compile ("[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
cnt = 0
for value in df.Email:
    if re.search(regex, value):
        pass
    else:
        df.loc[cnt, 'Email'] = np.nan
    cnt+=1
```

```python
cnt = 0
for value in df.PO_BOX:
    if len(str(value)) != 5 or (any (c.isalpha() for c in value)):
        df.loc[cnt, 'PO_BOX'] = np.nan
    cnt+=1
```
